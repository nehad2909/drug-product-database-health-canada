```python
#Import the necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import shutil
```

```python
#Create folders

cwd = os.getcwd()
#print(cwd)

dir = os.path.join(cwd, "Temp")
if not os.path.exists(dir):
    os.mkdir(dir)
dir = os.path.join(cwd, "DPD_Files")
if not os.path.exists(dir):
    os.mkdir(dir)
# dir = os.path.join(cwd, "DPD_Merged_Files")
# if not os.path.exists(dir):
#     os.mkdir(dir)
```

```python
drug = pd.read_csv('DPD_Files/drug.txt', header = None)
drug_ap = pd.read_csv('DPD_Files/drug_ap.txt', header = None)
drug_dr = pd.read_csv('DPD_Files/drug_dr.txt', header = None)
drug_ia = pd.read_csv('DPD_Files/drug_ia.txt', header = None)
```

```python
drug = [drug, drug_ap, drug_dr, drug_ia]
drug = pd.concat(drug)
#drug.to_csv('DPD_Merged_Files/drug.csv', index = False)
```
